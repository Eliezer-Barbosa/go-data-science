<h1><center> Visualização sobre um dataset de vendas de video game </center></h1>

```python
# definindo que a apresentação vai ser nesta aba
%matplotlib inline
# importando as libs necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# definindo o estilo dos gráficos
plt.style.use('ggplot')
```

```python
# leitura do dataset
videogames = pd.read_csv('dataset/vgsales.csv')
```

```python
# exibição dos 10 primeiros dados 
videogames.head(10)
```

```python
# apresentando uma breve descrição sobre os dados do dataset
videogames.describe()
```

```python
# identificando os tipos de cada coluna
videogames.dtypes
```

```python
videogames.shape
```

```python
#Renomeando colunas
videogames.columns = ['Ranking', 'Nome', 'Plataforma', 'Ano', 'Gênero', 'Editora', 'Vendas América do Norte', 'Vendas EUA', 'Vendas Japão', 'Outras vendas', 'Vendas Global']
videogames.head(10)
```

```python
# verificando campos com valores null
videogames[videogames['Ano'].isnull()].head(10)
```

```python
# somando os valores da coluna plataforma
titulos_lancados = videogames['Plataforma'].value_counts()
# criando um gráfico simples para apresentação dos resultados
titulos_lancados.plot()

videogames['Plataforma'].value_counts().plot()
```

```python
# uma melhor apresentação dos resultados
titulos_lancados = videogames['Plataforma'].value_counts().head(10).plot(kind='bar', figsize=(11, 5), grid=True, color='green')
# deixando o gráfico mais informativo
plt.title('Os 10 video games com mais titulos lançados')
plt.xlabel('Plataforma')
plt.ylabel('Quantidade')
plt.show()
```

```python
# intervalo de anos que existe no dataset
min(videogames['Ano']), max(videogames['Ano'])
```

```python
# cruzando valores sobre a Plataforma e Gênero
crosstab_vg = pd.crosstab(videogames['Plataforma'], videogames['Gênero'])
crosstab_vg.head()
```

```python
# adicionando a coluna total
crosstab_vg['Total'] = crosstab_vg.sum(axis=1)
crosstab_vg.head()
```

```python
# criando um heatmap para quantidade de titulos por gênero e console
top10_platforms = crosstab_vg[crosstab_vg['Total'] > 1000].sort_values('Total', ascending = False)
top10_final = top10_platforms.append(pd.DataFrame(top10_platforms.sum(), columns=['total']).T, ignore_index=False)
sns.set(font_scale=1)
plt.figure(figsize=(18, 9))
sns.heatmap(top10_final, annot=True, vmax=top10_final.loc[:'PS', :'Strategy'].values.max(), vmin=top10_final.loc[:, :'Strategy'].values.min(), fmt='d')
plt.title('Quantidade de titulos por gênero e plataforma')
plt.xlabel('Plataforma')
plt.ylabel('Gênero')
plt.show()
```
